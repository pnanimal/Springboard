```python
cproperties=properties.transpose() #transpose to get time data on x axis
cproperties=cproperties.reset_index() #move Buroughs out of index into df proper
cproperties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
cproperties.rename(columns={'Unnamed: 0' : 'Burough', pd.NaT : 'ID'}, inplace=True) #rename first two columns
cproperties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
dproperties=pd.melt(cproperties, id_vars=['Burough', 'ID']) #melt the data to make it tall, rather than long
dproperties.head(10)
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
dproperties.rename(columns={0 :'Date', 'value':'Average_prices'}, inplace=True)
dproperties['Average_prices']=pd.to_numeric(dproperties['Average_prices'], downcast='float') #cast avg prices to float
dproperties.dtypes #confirm that average prices is a float
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
dproperties['Burough'].nunique() #show that there are 48 instead of 32 Buroughs
```

```python
dproperties['Burough'].unique() #show that odd values include Unnamed: 47, Unnamed 37 in Buroughs list
```

```python
dproperties['Burough']=dproperties['Burough'].replace('Unnamed: 47', np.nan).replace('Unnamed: 37', np.nan)
eproperties=dproperties[dproperties['Burough'].notna()]
eproperties['Burough'].unique() #these values have been dropped
```

```python
eproperties['ID'].unique() #show that there are nans in the ID also
```

```python
fproperties=eproperties[eproperties['ID'].replace('nan',np.nan).notna()]
fproperties['ID'].unique() #nans have been removed
```

```python
fproperties['Burough'].nunique() #there are still too many Buroughs (need 32)
```

```python
fproperties['Burough'].unique() #some of them are in all caps 
```

```python
fproperties['bad_data']=[x.isupper() for x in fproperties['Burough']] #check which Buroughs are all CAPS
gproperties = fproperties[fproperties.bad_data !=1].drop('bad_data', axis=1) #remove the all CAPS Buroughs
gproperties.head()
```

```python
gproperties['Burough'].nunique() #But there are still 4 too many Buroughs
```

```python
gproperties['Burough'].unique() #The following look invalid: City of London, Inner London, Outer London and England
```

```python
more_bad_data = ['City of London','Inner London','Outer London','England'] #make a list of the additional suspect values
gproperties=gproperties[~gproperties['Burough'].isin(more_bad_data)]
Blist=gproperties['Burough'].unique()
print(Blist)
```

```python
gproperties['Burough'].nunique() #now there are only 32 Buroughs
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
xval = gproperties[gproperties['Burough']=='Camden']['Date']
yval = gproperties[gproperties['Burough']=='Camden']['Average_prices']
plt.plot(xval,yval/1000)
plt.title('Average price over time for Camden Burough')
plt.xlabel('Date')
plt.ylabel('Average price ($1000)')
plt.show
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Create a new dataframe of price ratios using a list in a loop
PriceList=[]
for i in Blist: #list of the buroughs
    val1998 = gproperties[(gproperties.Burough==i) & (gproperties.year==1998)].Average_prices.mean()
    val2019 = gproperties[(gproperties.Burough==i) & (gproperties.year==2019)].Average_prices.mean()
    Prices = [i, val1998, val2019, val1998/val2019]
    PriceList.append(Prices)
AvgPricedf = pd.DataFrame(PriceList, columns = ['Burough','1998 Avg Price','2019 Avg Price','Price Ratio']) 
AvgPricedf.head(20)
```

```python
for i in Blist:
    xval = AvgPricedf[AvgPricedf['Burough']==i]['1998 Avg Price']
    yval = AvgPricedf[AvgPricedf['Burough']==i]['2019 Avg Price']
    plt.scatter(xval, yval, s=pow(1*(yval/xval),3.5), alpha=0.6) #note that scale is computed by 2019/1998 to show increase
plt.legend(Blist,loc='upper center', bbox_to_anchor=(1.45, 0.8))
plt.title('Average Prices London Buroughs, scaled by price increase')
plt.xlabel('1998 Price')
plt.ylabel('2019 Price')
plt.show
```

```python
#Look at the 5 highest average priced Buroughs, which can be isolated using 1998 prices
AvgPricedf_sorted=AvgPricedf.sort_values(by='1998 Avg Price', ascending=False) 
AvgPricedf_sorted.head(10)
```

```python
#Take the top 5 buroughs
HighAvgPricedf=AvgPricedf_sorted[:5]
Clist=HighAvgPricedf['Burough'].unique()
for i in Clist:
    xval = HighAvgPricedf[HighAvgPricedf['Burough']==i]['1998 Avg Price']
    yval = HighAvgPricedf[HighAvgPricedf['Burough']==i]['2019 Avg Price']
    plt.scatter(xval, yval, s=pow(1*(yval/xval),3.5), alpha=0.6)
plt.legend(Clist,loc='upper center', bbox_to_anchor=(1.45, 0.8))
plt.title('Average Prices top 5 London Buroughs, scaled by price increase')
plt.xlabel('1998 Price')
plt.ylabel('2019 Price')
plt.show
```

```python
#What are the Buroughs with the highest prices increases (lowest price ratios)?
Ratiodf_sorted=AvgPricedf.sort_values(by='Price Ratio', ascending=True) 
Ratiodf_sorted.head(10)
```

```python
#Take the 5 buroughs with the highest price increases
HighIncreasedf=Ratiodf_sorted[:5]
Dlist=HighIncreasedf['Burough'].unique()
for i in Dlist:
    xval = HighIncreasedf[HighIncreasedf['Burough']==i]['1998 Avg Price']
    yval = HighIncreasedf[HighIncreasedf['Burough']==i]['2019 Avg Price']
    plt.scatter(xval, yval, s=pow(1*(yval/xval),3), alpha=0.6)
plt.legend(Dlist,loc='upper center', bbox_to_anchor=(1.45, 0.8))
plt.title('Top 5 prices increases London Buroughs, 1998 to 2019')
plt.xlabel('1998 Price')
plt.ylabel('2019 Price')
plt.show
```
